```python
# Get the tweets required for the analysis
query = 'Machine Learning'
public_tweets = api.search(q=[query], count=100)

count_neutral = 0
count_positive = 0
count_negative = 0
tweet_dict ={}

# perform the sentiment analysis

for tweet in public_tweets:
    analysis = TextBlob(tweet.text)
    polarity_val = analysis.sentiment.polarity

    if (polarity_val == 0.0):
        print (tweet.text)
        count_neutral += 1
        tweet_dict[tweet.text.encode('utf8')] = 'Neutral'
    elif (polarity_val < 0.0):
        print ('\x1b[1;31m'+ tweet.text +'\x1b[0m')
        count_negative += 1
        tweet_dict[tweet.text.encode('utf8')] = 'Negative'
    else:
        print('\x1b[1;32m'+ tweet.text +'\x1b[0m')
        count_positive += 1
        tweet_dict[tweet.text.encode('utf8')] = 'Positive'
    print("\n")
    
print ("Neutral Count = %d" % count_neutral)    
print ("Positive Count = %d" % count_positive)
print ("Negative Count = %d" % count_negative)
```
