```python
public_tweets = api.search('Google', count=100)

count_neutral = 0
count_positive = 0
count_negative = 0

for tweet in public_tweets:
    print(tweet.text)
    analysis = TextBlob(tweet.text)
    polarity_val = analysis.sentiment.polarity

    if (polarity_val == 0.0):
        print ("Neutral")
        count_neutral += 1
    elif (polarity_val < 0.0):
        print ('\x1b[1;31m'+ 'Negative' +'\x1b[0m')
        count_negative += 1
    else:
        print('\x1b[1;32m'+ 'Positive' +'\x1b[0m')
        count_positive += 1
    print("\n")
    
print ("Neutral Count = %d" % count_neutral)    
print ("Positive Count = %d" % count_positive)
print ("Negative Count = %d" % count_negative)
```
